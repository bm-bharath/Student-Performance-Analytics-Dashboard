```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../processed/student_performance_clean.csv")
df.head()
```

```python
df.info()
df.describe()

# Distribution of ExamScore
sns.histplot(df["ExamScore"], kde=True)
plt.title("Exam Score Distribution")
plt.show()

# Attendance vs ExamScore
sns.scatterplot(data=df, x="Attendance", y="ExamScore", hue="PerformanceCategory")
plt.title("Attendance vs Exam Score")
plt.show()

# Learning Style vs ExamScore
sns.boxplot(data=df, x="LearningStyle", y="ExamScore")
plt.title("Exam Score by Learning Style")
plt.xticks(rotation=45)
plt.show()
```
